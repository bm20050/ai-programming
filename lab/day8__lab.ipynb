```python
import numpy as np

def print_list(lst, labels=None):
  import string
  labels = string.ascii_lowercase[:len(lst)+1] if labels is None else labels
  for label, data in list(zip(labels, lst)):
    print(label, data, sep='\n')

a = np.array ([-1.7, 1.5, -0.2, 0.6, 10])
b = np.around(a)
c = np.floor(a)
d = np.ceil(a)
print_list([a, b, c, d])

a = np.array([1.0, 5.55, 123, 0.567, 25.532])
b = np.around(a)
c = np.around(a, decimals = 1)
d = np.around(a, decimals = -1) # default is 0
print_list([a, b, c, d])
```

## Statistical Functions

```python
import numpy as np

a = np.array([[3,7,5],[8,4,3],[2,4,9]])
b = np.amin(a, 1) # Minimum element along the rows (axis=1)
c = np.amin(a, 0) # Minimum element along the cols (axis=0)
d = np.amin(a) # Minimum element in the given array
e = np.amax(a, 1)
f = np.amax(a, 0)

a = np.array([1,2,3,4])
g = np.average(a)
h = np.std(a) # Standard deviation
i = np.var(a) # Variance

print_list([b, c, d, e, f, g, h, i])
```

## p.17 Linear Algebra




```python
import numpy as np

def examples():
  a = np.array([[1,2],[3,4]])
  b = np.array([[-4],[1]])
  inv_a = np.linalg.inv(a)  # inverse of a
  c = np.linalg.solve(a, b) # compute x such that ax=b
  d = np.linalg.det(a) # determinant of a
  e = np.linalg.eig(a) # eigenvalues and eigenvectors of a
  print_list([a, b, inv_a, c, d, e], locals().keys())

examples()
```

```python
import numpy as np
import matplotlib.pyplot as plt
fileName = '/content/gdrive/MyDrive/firstChoice.txt'
def read_data(fileName):
  result = list()
  with open(fileName, 'r') as f_read:
    for line in f_read:
      result.append(float(line))
  return result

# Search performance by First-choice hill climbing for TSP
x = np.arange(14140) # There are 14140 values to be plotted
y = read_data(fileName) # Y values will be read in from a file

plt.plot(x, y)
plt.xlabel('Number of Evaluations')
plt.ylabel('Tour Cost')
plt.title('TSP-100 by First-Choice Hill Climbing')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Plot for a Gaussian function
def gaussian(x):
  return np.exp(-(0.5 - x)**2/1.5)
  
x = np.arange(-2, 2.5, 0.01)
y = gaussian(x) # Array argument produces array output
plt.plot(x, y)
plt.xlabel('x values')
plt.ylabel('exp(-(0.5-x)**2/1.5)')
plt.title('Gaussian Function')
plt.show()
```
